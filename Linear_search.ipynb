```python
import matplotlib.pyplot as plt

def linearSearch(arr,key,n)->int:
  for i in range(0,n+1):
    if(arr[i]==key):
      return i

arr=[]
for i in range(0, 101):
    arr.append(i*10)

best=[]
average=[]
worst=[]
arrsize=[]

def main_plot():
    for n in range(10,101,5):
        arrsize.append(n)
        key_best=arr[0]
        key_average=arr[int(n/2)]
        key_worst=arr[n-1]

        best.append(linearSearch(arr,key_best,n)+1)
        average.append(linearSearch(arr,key_average,n))
        worst.append(linearSearch(arr,key_worst,n)+1)
    

def key_Search():
    for n in range(10,101,5):
        for i in range(0,n+1):
            if i<n:
                found=linearSearch(arr,i,n)
                print("Successfull key Search, element found at index: ", i)
            else:
                print("Unsuccessfull key Search\n",)
                
main_plot() #Plot  the  graph for each case -  best, worst and average number of comparisons. n varies from 10 to 100 in steps of 5

key_Search() #For every input size n , run it with n+ 1 different keys - n successful plus 1 unsuccessful
```
